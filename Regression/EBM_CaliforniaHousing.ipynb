```python
from piml import Experiment
exp = Experiment()
```

```python
# Choose CaliforniaHousing_trim2
exp.data_loader()
```

```python
exp.data_summary()
```

```python
# Prepare dataset with default settings
exp.data_prepare()
```

```python
# Optional: feature selection by LGBM feature importance or RCIT Conditional Independence Test 
# Try: LGBM, choose Top Percentage 98%
exp.feature_select()
```

```python
# Exploratory data analysis, check distribution and correlation
exp.eda()
```

```python
# First, choose GLM and ReLU-DNN with default settings, click run;
# Then, choose only ReLU-DNN and customize it with L1=0.0005; Reigster the three models
exp.model_train()
```

```python
# Model-specific inherent interpretability: feature importance, main effects and pairwise interactions
exp.model_interpret()
```

```python
# Model-agnostic post-hoc explanability: global methods (PFI, PDP, ALE) and local methods (LIME, SHAP)
exp.model_explain()
```

```python
exp.model_diagnose()
```

```python
exp.model_compare()
```
