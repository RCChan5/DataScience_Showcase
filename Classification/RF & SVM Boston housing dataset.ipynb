```python
import pandas as pd
import numpy as np

data= pd.read_csv("BostonHousing.csv")
data.describe().T

```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import StandardScaler



y = data['CAT. MEDV']
X = data.drop(['CAT. MEDV','MEDV'],axis=1)
#X = X.drop('MEDV',axis=1)

#from sklearn.preprocessing import StandardScaler

#X=StandardScaler().fit_transform(X)
X_train,X_val,y_train,y_val = train_test_split(X,y,test_size=0.3,random_state=0)

print(type(data))
```

```python
X.describe().T
```

```python
print(X)
```

```python
data.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0) # n_estimator = 100 (default);␣,→default=”sqrt”


#cross validation of the training data set since an acuracy of 100% seemed alittle off
cv = cross_validate(rf, X, y, cv=5)


print("kfold cross validation")
print(cv['test_score'])
print(cv['test_score'].mean())



rf.fit(X_train,y_train)
```

```python
rf.score(X_val,y_val)
```

```python
X.shape
```

```python
from sklearn.model_selection import GridSearchCV # parameter tuning
import numpy as np
param = {
'n_estimators': [50,100], # number of trees
'max_features': range(1,13) # number of mtrees
}
rf2 = RandomForestClassifier(random_state=0)
grid_search = GridSearchCV(estimator=rf2,
param_grid=param, n_jobs=-1) # Number of jobs to run␣in parallel. -1 means using all
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.score(X_val,y_val)
```

```python
final_model = grid_search.best_estimator_
# feature importance
feature_imp = pd.Series(final_model.feature_importances_,index=X.columns).sort_values(ascending=False)
feature_imp
```

```python

rf2.fit(X_train,y_train)
# model performance -- confusion matrix
from sklearn.metrics import confusion_matrix
prediction = rf2.predict(X_val)
confusion_matrix(y_val,prediction)
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix
print("Tuned Random Forest")
y_pred = prediction

# Calculate accuracy, sensitivity, and specificity
accuracy = accuracy_score(y_val, y_pred)
tn, fp, fn, tp = confusion_matrix(y_val, y_pred).ravel()
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

# Print the results
print(f"Accuracy: {accuracy:.3f}")
print(f"Sensitivity: {sensitivity:.3f}")
print(f"Specificity: {specificity:.3f}")
```
