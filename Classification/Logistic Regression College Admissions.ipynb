```python
# Data partation
y=dataRC['Enroll']
from sklearn.model_selection import train_test_split
X_trainRC,X_valRC,y_trainRC,y_valRC=train_test_split(XRC,y,test_size=0.3,random_state=0)
```

```python
# Fit a logistic regression
import statsmodels.api as sm
log_regRC = sm.Logit(y_trainRC,X_trainRC).fit()
print(log_regRC.summary())
```

```python
 # 10. Model Evaluation (accuracy + confusion matrix)
from sklearn.metrics import confusion_matrix, accuracy_score,ConfusionMatrixDisplay
prediction_probRC =log_regRC.predict(X_valRC)
predictionRC =list(map(round,prediction_probRC))
cm = confusion_matrix(y_valRC,predictionRC)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
sensitivity = cm[1, 1] / (cm[1, 1] + cm[1, 0])
print('Sensitivity: ', sensitivity)
print("Accuracy Score")
print(accuracy_score(y_valRC,predictionRC))

print("specificity")
#tn / (tn+fp)
print(26417/(26417+181))
```

```python
# Calculate sensitivity (True Positive Rate)
sensitivity = cm[1, 1] / (cm[1, 1] + cm[1, 0])
print('Sensitivity: ', sensitivity)

# Calculate specificity (True Negative Rate)
specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1])
print('Specificity: ', specificity)

# Calculate accuracy
accuracy = np.sum(np.diag(cm)) / np.sum(cm)
print('Accuracy: ', accuracy)
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
from matplotlib import pyplot as plt 

lr_aucRC = roc_auc_score(y_valRC,prediction_probRC)
print('Logistic: ROC AUC =%.3f' % (lr_aucRC))
lr_fprRC, lr_tprRC, threshold = roc_curve(y_valRC,prediction_probRC)
plt.plot(lr_fprRC,lr_tprRC,marker='.')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```
