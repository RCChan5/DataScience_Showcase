```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dataRC= pd.read_csv("inq2015.csv")
dataRC.describe()
```

```python
# Calculate the percentage of missing values for each column
missing_values = dataRC.isnull().mean() * 100

#Print the missing value percentage for each column
print(missing_values)
```

```python
dataRC=dataRC.drop(columns=["ACADEMIC_INTEREST_1","ACADEMIC_INTEREST_2","IRSCHOOL","CONTACT_CODE1","CONTACT_DATE"])
#Drop LEVEL YEAR nunique is only 1 meaning its the same for all of the samples
dataRC=dataRC.drop(columns=["LEVEL_YEAR"])
#drop territory no meta dataRC was provided I dont know what the points relate to and the nomenclature within the variable(N and A)
#keep it for now 
#dataRC=dataRC.drop(columns=["TERRITORY"])
#drop satscore avg income and distance too many missing values

dataRC=dataRC.drop(columns=["satscore"])
dataRC=dataRC.drop(columns=["telecq"])

dataRC = dataRC[dataRC['TERRITORY'].notna()]
```

```python
# Calculate the percentage of missing values for each column
missing_values = dataRC.isnull().mean() * 100

#Print the missing value percentage for each column
print(missing_values)
```

```python
#print(dataRC.ETHNICITY.mode())
dataRC["ETHNICITY"]=dataRC["ETHNICITY"].fillna("C")
#print(dataRC.sex.mode())
dataRC["sex"]=dataRC["sex"].fillna(1)
#print(dataRC.distance.mean())
dataRC["distance"]=dataRC["distance"].fillna(380.4276)
#print(dataRC.avg_income.mean())
dataRC["avg_income"]=dataRC["avg_income"].fillna(47315.334)
```

```python
# Calculate the percentage of missing values for each column
missing_values = dataRC.isnull().mean() * 100

#Print the missing value percentage for each column
print(missing_values)
```

```python
dummy_dataRC =pd.get_dummies(dataRC[["ETHNICITY","Instate","TERRITORY"]])

data_contRC = dataRC[["SELF_INIT_CNTCTS","TRAVEL_INIT_CNTCTS", "CAMPUS_VISIT", "interest", "init_span", "int1rat","int2rat","avg_income","distance", "hscrat","sex","mailq","premiere","stucell"]]
dummy_dataRC.describe()
```

```python
num_namesRC= ["SELF_INIT_CNTCTS","TRAVEL_INIT_CNTCTS", "CAMPUS_VISIT", "interest", "init_span", "int1rat","int2rat","avg_income","distance", "hscrat","sex","mailq","premiere","stucell"]
dum_namesRC= ["ETHNICITY_A","ETHNICITY_B","ETHNICITY_C","ETHNICITY_H","ETHNICITY_O","ETHNICITY_N","Instate_N","TERRITORY_N","TERRITORY_1","TERRITORY_2","TERRITORY_3","TERRITORY_4","TERRITORY_5","TERRITORY_6","TERRITORY_7","TERRITORY_8","TERRITORY_A"]
XRC = pd.concat([dataRC[num_namesRC],dummy_dataRC[dum_namesRC]],axis=1) 
XRC.describe(include='all')
```

```python
y = dataRC['Enroll']
from sklearn.model_selection import train_test_split
X_train, X_val, y_train,y_val =train_test_split(XRC,y,test_size=0.3,random_state=0)
```

```python
from sklearn import tree
from sklearn.tree import export_text
dtree = tree.DecisionTreeClassifier(max_depth=4, min_samples_split=30,random_state = 0)
dtree = dtree.fit(X_train,y_train)
```

```python
r = export_text(dtree, feature_names=list(X_train.columns.values))
print(r)
```

```python
# visualization of decision tree
from matplotlib import pyplot as plt
plt.figure(figsize=[25,20])
tree.plot_tree(dtree,
feature_names=list(X_train.columns.values),
class_names=True,
filled=True)
plt.show()
```

```python

# Extract feature importances from the model
importances = dtree.feature_importances_

# Create a dataframe to view feature importances
importance_df = pd.DataFrame({'feature': XRC.columns, 'importance': importances})
importance_df = importance_df.sort_values('importance', ascending=False)

# Print feature importances
print(importance_df)
```

```python
# model performance -- confusion matrix
from sklearn.metrics import confusion_matrix
prediction = dtree.predict(X_val)
cm = confusion_matrix(y_val,prediction)
sensitivity = cm[1, 1] / (cm[1, 1] + cm[1, 0])
print('Sensitivity: ', sensitivity)
```

```python
print("Accuracy")
dtree.score(X_val,y_val) # accuracy
```

```python
 # other measures
tn, fp, fn, tp = confusion_matrix(y_val, prediction).ravel()
specificity = tn / (tn+fp)

print("Specificity")
print(specificity)
```

```python
# Calculate sensitivity (True Positive Rate)
sensitivity = cm[1, 1] / (cm[1, 1] + cm[1, 0])
print('Sensitivity: ', sensitivity)

# Calculate specificity (True Negative Rate)
specificity = cm[0, 0] / (cm[0, 0] + cm[0, 1])
print('Specificity: ', specificity)

# Calculate accuracy
accuracy = np.sum(np.diag(cm)) / np.sum(cm)
print('Accuracy: ', accuracy)
```

```python
# roc curve and auc score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
predic_prob =dtree.predict_proba(X_val) # Probability estimates
print(predic_prob)
tree_auc =roc_auc_score(y_val, predic_prob[:,1])
print(tree_auc)
```

```python
# visualization of roc curve
tree_fpr, tree_tpr, threshold =roc_curve(y_val, predic_prob[:,1])
plt.plot(tree_fpr,tree_tpr,marker='.')
plt.xlabel('False Positive Rate')
plt.ylabel('Truee Positive Rate')
plt.show()
```

```python
# tree 2 pruning
from sklearn import tree
from sklearn.tree import export_text
dtree2 = tree.DecisionTreeClassifier()
dtree2 = dtree2.fit(X_train,y_train)
```

```python
r2 = export_text(dtree2, feature_names=list(X_train.columns.values))
#print(r2)
```

```python
 # pruning the tree
path = dtree2.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
dtree2s = []
for ccp_alpha in ccp_alphas:
    dtree2 = tree.DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    dtree2.fit(X_train, y_train)
    dtree2s.append(dtree2)
print(
"Number of nodes in the last tree is: {} with ccp_alpha: {}".format(
dtree2s[-1].tree_.node_count, ccp_alphas[-1]
)
)
```

```python
# model evaluation based on the validation data set
train_scores = [dtree2.score(X_train, y_train) for dtree2 in dtree2s]
validation_scores = [dtree2.score(X_val, y_val) for dtree2 in dtree2s]
print(validation_scores)
print(ccp_alphas)
accuracy_list = pd.DataFrame(
{'validation_scores': validation_scores,
'ccp_alphas': ccp_alphas,
})
accuracy_list

print(max(accuracy_list["validation_scores"]))
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and validation sets")
ax.plot(ccp_alphas, train_scores, marker="o", label="train",drawstyle="steps-post")
ax.plot(ccp_alphas, validation_scores, marker="o", label="validation",drawstyle="steps-post")
ax.legend()
plt.show()
```
